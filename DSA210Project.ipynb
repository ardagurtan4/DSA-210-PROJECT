```python

import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Date": ["2025-01-07", "2025-01-08", "2025-01-09"],
    "Studying Periods (Hours)": [4.36, 3.15, 1.48]  
}

df = pd.DataFrame(data)

print("Studying Hours Table:")
print(df)

plt.figure(figsize=(8, 5))
plt.bar(df["Date"], df["Studying Periods (Hours)"])
plt.title("Studying Hours (7-9 January 2025)")
plt.xlabel("Date")
plt.ylabel("Studying Periods (Hours)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
